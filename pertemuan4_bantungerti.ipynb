### 2. traffic source apa yang paling banyak datangin pelanggan per bulan

```python
df.columns
### select month,traffic source,count(*) from df gorup by month, traffic
```

```python
df.info()
```

```python
df['order_date']=pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
df['month']=df['order_date'].dt.month
print(df['month'].unique())
```

```python
traffic_per_month=df.groupby(['month','traffic_source'])['id'].nunique().unstack()
### select count(id),'month','traffic_source' form df group by 'month','traffic_source'
```

```python
traffic_per_month.plot(kind='line')
plt.xlabel('month')
plt.ylabel('jumlah pelanggan')
plt.title('pergerakan jumlah traffic per bulan')
```

### 3. Ada berapa total jumlah customer perbulan ini

```python

total_customers=df.groupby('month')['id'].nunique()
total_customers.plot(kind='line')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Customer')
plt.title('Total Customer per Bulan')
```

### 4. Total jumlah order perbulan yang complete

```python
dfComplete=df[df['status_order']=='Complete']
```

```python
jumlahOrderBerhasil=dfComplete.groupby('month')['id'].nunique()
```

```python
jumlahOrderBerhasil.plot(kind='line')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Order')
plt.title('Total Order Berhasil per Bulan')
```

### 5. Ada berapa status order dataset

```python
dfStatusOrder=df.groupby('status_order')['id'].nunique()
```

```python
dfStatusOrder.plot(kind='bar')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Order')
plt.title('Total Order Berhasil per Bulan')
```
